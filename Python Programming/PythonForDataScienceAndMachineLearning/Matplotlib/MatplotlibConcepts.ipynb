```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
## Basic Matplotlib Commands
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

Creating Multiplots on Same Canvas

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'b')
plt.show()
```

Matplotlib Object Oriented Method

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
plt.show()
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
plt.show()
```

Figure, Size and DPI

```python
fig = plt.figure(figsize=(8,2), dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
#Setting Labels on figures
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0,1,1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.plot(y, x, 'r')
axes.set_xlabel('X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Y Label')
axes.set_title('Title')
plt.show()
plt.tight_layout()
```

```python
#Setting Labels on figures
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0,1,1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, x**2,label="X Squared") #Adding legend titles
axes.plot(x, x**3, label="X Cubed")  #Adding legend titles
axes.legend() #Adding labels to Legend

axes.set_xlabel('X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Y Label')
axes.set_title('Title')

plt.show()
plt.tight_layout()
```

Plot Appearance

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, y ,color='green') #Can also use RBG Hex Code
plt.show()
```

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, y ,color='purple', linewidth=3) #Add linewidth
plt.show()
```

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, y ,color='purple', linewidth=5, alpha=0.5) #Add transparency of line
plt.show()
```

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, y ,color='purple', lw=5, linestyle='-.') #Add line style
plt.show()
```

```python
# MATLAB style line color and style 
plt.close('all') 
fig, ax = plt.subplots()
ax.plot(
    x, y,
    color='orange',     # line color
    lw=5,               # line width
    ls='--',            # line style
    marker='o',         # marker shape
    markersize=10,      # marker size
    markeredgecolor='blue',  # marker border color
    markeredgewidth=4   # marker border thickness
)
plt.show()
```

Plot Range

```python
# PLOT RANGE
plt.close('all') 
fig, ax = plt.subplots()
ax.plot( x, y,color='blue',lw=5, ls='--')
ax.set_xlim([0, 1]) #Set x limit (Range)
ax.set_ylim([0, 2])  #Set y limit (Range)
plt.show()
```
